# JSON

Since its creating, JSON (Javascript Object Notation) is the standard way to exchange data within the internet or data bases.

## Structure

The structure of the JSON is nothing more than a key-value pair:

key: a string with the name of the field
value: any object related to the key, as ints, floats, strings, lists, null etc.

## Using JSON with Python

Desserializing JSON files with **json.load**

```python
archive = open('./archives/JsonExample.txt', encoding='utf8')

json_archive = json.load(archive)

archive.close()

print(type(json_archive))
print(json_archive)
```

You can serialize again the json object considered as a dict with **json.dumps(deserialized_archive, ensure_ascii=False)**.

Now, let's discover how to serialize a disctionary on python into a json file (.json)

```python
filme = {}
filme['Nome'] = 'Toy Story'
filme['Direção'] = 'John Lasseter'
filme['Produção executiva'] = ['Edwin Catmull','Steve Jobs']
filme['Gênero'] = ['animação','aventura','comédia']
filme['Produtora'] = ['Walt Disney Pictures','Pixar Animation Studios']
filme['Lançamento'] = [{'País': 'Estados Unidos',
                      'Data': '22/11/1995'},
                      {'País': 'Brasil',
                      'Data': '22/12/1995'},
                      {'País': 'Portugal',
                      'Data': '29/03/1996'}]
filme['Orçamento (US$)'] = 30000000
filme['Receita (US$)'] = 406594102
 
arquivo_json = open('./archives/JSONSerialization.json',mode='w',encoding='utf8')
 
json.dump(filme, arquivo_json, ensure_ascii=False)
 
arquivo_json.close()
```
